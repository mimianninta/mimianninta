```
import pandas as pd
df = pd.read_csv("a.csv")
df.head(5)
```

```
df["ImageID"][0]
```

```
length = df.shape[0]
length
```

```
import os

tr = []
vl = []
f1 = open("train.txt","r")
f2 = open("valid.txt","r")
l1 = f1.readlines()
l2 = f2.readlines()

for x in l1:
  xx = x.rstrip()
  xx = xx[31:]
  xx = xx[:-4]
  tr.append(xx)

for x in l2:
  xx = x.rstrip()
  xx = xx[31:]
  xx = xx[:-4]
  vl.append(xx)
       
f1.close()
f2.close()
vl
```

```
a=[]
ll=[]
for i in range(length):
  if df["ImageID"][i] in tr:
    ll = [df["ImageID"][i], df["XMin"][i], df["XMax"][i], df["YMin"][i], df["YMax"][i]]
    a.append(ll)

trn = pd.DataFrame(a, columns =['ImageID',	'XMin',	'XMax',	'YMin',	'YMax']) 
trn
```

```
trn.to_csv("fish_train.csv")
```

```
a=[]
ll=[]
for i in range(length):
  if df["ImageID"][i] in vl:
    ll = [df["ImageID"][i], df["XMin"][i], df["XMax"][i], df["YMin"][i], df["YMax"][i]]
    a.append(ll)

val = pd.DataFrame(a, columns =['ImageID',	'XMin',	'XMax',	'YMin',	'YMax']) 
val
```

```
val.to_csv("fish_val.csv")
```
